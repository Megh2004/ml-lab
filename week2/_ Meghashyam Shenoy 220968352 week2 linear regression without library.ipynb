```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
df=pd.read_csv("d1.csv")
df
```

```python
plt.scatter(x=df['x'],y=df['y'])
for x, y in zip(df['x'], df['y']):
    plt.text(x, y, f'({x}, {y})')

plt.title("sales vs number of weeks")
plt.xlabel("number of weeks")
plt.ylabel("Sales in thousands")
```

```python
x=df['x'].to_numpy()
y=df['y'].to_numpy()
```

```python
a=np.mean(x)
```

```python
b=np.mean(y)
print(b)
```

```python
v=np.var(x)
```

```python
r=np.multiply(x,y)
s=r.mean()
print(s)
```

```python
s=s-a*b
```

```python
a1=s/v
```

```python
print(a1)
```

```python
a0=b-a1*a
print(a0)
```

```python
df['y1']=a0+a1*df['x']
```

```python
plt.plot(x, a0+a1*x, color="red")
for x, y in zip(df['x'], df['y']):
    plt.text(x, y, f'({x}, {y})')

plt.scatter(x=df['x'],y=df['y'])
plt.title("sales vs number of weeks")
plt.xlabel("number of weeks")
plt.ylabel("Sales in thousands")
plt.show()
```

```python
#for the 7th and 12th week
a7=a0+a1*7
a12=a0+a1*12
print(a7)
print(a12)
```

# linear regression for random integers  values

```python
c=np.array(random.sample(range(50),10))
d=np.array(random.sample(range(50),10))
```

```python
plt.scatter(x=c,y=d)
plt.title("random integers")
```

```python
e=np.mean(c)
f=np.mean(d)
g=np.var(c)
```

```python
h=np.multiply(c,d)
h=np.mean(h)
```

```python
i=h-e*f
a2=i/g
```

```python
print(a2)
```

```python
a1=f-a2*e
```

```python
print(a1)
```

```python
plt.plot(c, a1+a2*c, color="red")
plt.scatter(c,d)
plt.show()
```

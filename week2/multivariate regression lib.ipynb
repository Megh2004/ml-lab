```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("d2.csv")
df
```

```python
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['X1'], df['X2'], df['Y'])

ax.set_title("Y RELATIONSHIP WITH X1,X2")
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("Y")

plt.show()
```

#coefficient(x1,x1) and intercept

```python
X = df[['X1', 'X2']]  # Replace with your column names
y = df['Y'] 
model = LinearRegression()
model.fit(X, y)
print(model.coef_)
print(model.intercept_)
```

```python
print(model.predict([[3,2]]))
```

#for random integers

```python
X1 = np.random.randint(0,20,size=(10,1))  
X2=np.random.randint(0,20,size=(10,1))  
Y= np.random.randint(0,20,size=(10,1))
```

```python
ax=plt.figure().add_subplot(111,projection='3d')
ax.scatter(X1,X2,Y)
X=np.concatenate((X1,X2),axis=1)
```

```python
from sklearn.linear_model import LinearRegression
m=LinearRegression()
m.fit(X,Y)
```

```python
c=m.coef_
d=c[0,1]
e=c[0,0]
print(c[0,0])
```

```python
f=m.intercept_
print(f)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

ax=plt.figure().add_subplot(111,projection='3d')
ax.scatter(X1,X2,Y)

X1_plane, X2_plane = np.meshgrid(np.linspace(X1.min(), X1.max(), 10), np.linspace(X2.min(), X2.max(), 10))
Y_plane = f + e * X1_plane + d * X2_plane
ax.plot_surface(X1_plane, X2_plane, Y_plane, alpha=0.5, color='r', label='Regression Plane')
```

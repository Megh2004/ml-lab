```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("d1.csv")
df.describe()
df.isnull().sum()
```

```python
df.info()
df['x']=df['x'].astype(float)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=df['x'],y=df['y'])
for x, y in zip(df['x'], df['y']):
    plt.text(x, y, f'({x}, {y})')

plt.title("x vs y")
plt.xlabel("x")
plt.ylabel("Y")
plt.show()
```

```python

x=df[['x']].values
y=df['y'].values
```

```python
regr=LinearRegression()
regr.fit(x,y)
```

```python
print(regr.coef_) 
```

```python
print(regr.intercept_)
```

```python
#for the 7th and 12th value
print(regr.predict([[7]]))
print(regr.predict([[12]]))
```

# linear regression for random integers values

```python

X = np.random.randint(0,20,size=(10, 1))  


Y= np.random.randint(0,20,size=(10,1))
plt.figure(figsize=(10,6))
plt.scatter(x=X,y=Y)
for x, y in zip(X, Y):
    plt.text(x, y, f'({x}, {y})')
plt.title("x vs y")
plt.xlabel("x")
plt.ylabel("Y")
plt.show()
regr = LinearRegression()
regr.fit(X, Y)
```

```python
C=regr.coef_
print(regr.coef_) 
```

```python
D=regr.intercept_
print(regr.intercept_)
```

```python
plt.figure(figsize=(10,10))
plt.plot(X,C*X+D)

plt.scatter(X,Y)
for x,y in zip(X,Y):
    plt.text(x,y,f'({x},{y})')
plt.show()
```

# LR FOR random float

```python
X=np.random.uniform(0,50.1,size=(10,1))
Y=np.random.uniform(0,50.1,size=(10,1))
plt.figure(figsize=(10,6))
plt.scatter(x=X,y=Y)
plt.title("X VS Y")
plt.xlabel("x")
plt.ylabel("y")
for x, y in zip(X, Y):
    plt.text(x, y, f'({x}, {y})')
plt.show()
m=LinearRegression()
m.fit(X,Y)
```

```python
c=m.coef_
d=m.intercept_
print(c)
print(d)
```

```python
plt.plot(X,c*X+d)
plt.scatter(X,c*X+d)
for i,j in zip(X,Y):
    plt.text(i,j,f'({i},{j})')
plt.scatter(X,Y)
```

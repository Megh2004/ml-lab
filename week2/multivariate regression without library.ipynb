```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
df=pd.read_csv("d2.csv")
df
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['X1'], df['X2'], df['Y'])

ax.set_title("Y RELATIONSHIP WITH X1,X2")
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("Y")

plt.show()
```

```python
x1=df['X1'].to_numpy()
x2=df['X2'].to_numpy()
y=df['Y'].to_numpy()
```

```python
a=np.sum((np.multiply(x1,x2)))
print(a)
b=np.sum((np.multiply(x1,y)))
print(b)
c=np.sum(np.multiply(x1,x1))


print(c)
```

```python
d=np.sum(np.multiply(x2,y))
e=np.sum(np.multiply(x2,x2))
print(d,e)
```

```python
X1=c-((np.sum(x1))**2)/len(x1)
```

```python
print(X1)
```

```python
X2=e-((np.sum(x2))**2)/len(x2)
print(X2)
```

```python
X1Y=b-(np.sum(x1)*np.sum(y))/5
X2Y=d-(np.sum(x2)*np.sum(y))/5
X1X2=a-(np.sum(x1)*np.sum(x2)/5)
print(X1Y,X2Y,X1X2)
```

```python
de=X1*X2-X1X2**2
print(de)
```

```python
b1=(X2*X1Y-X1X2*X2Y)/de
print(b1)
```

```python
b2=(X1*X2Y-X1X2*X1Y)/de
print(b2)
```

```python
np.mean(x1)
```

```python
a=np.mean(y)-b1*np.mean(x1)-b2*np.mean(x2)
```

```python
print(a)
```

# Final prediction

```python
y=a+b1*3+b2*2
print(y)
```
